#matplotlibグラフ表示のライブラリ

```python
#読み込み命令
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

###グラフ出力するデータの準備
###グラフのラベルとグリッドを描画させる
###出来上がったグラフをpyplotで表示する

```python
t=np.arange(0.0,2.0,0.01)#0.0から2.0まで0.01刻みでデータを生成
#SINカーブを描くために、元データのsin値を取得して変数sに設定
s=1+np.sin(2*np.pi*t)

#グラフの描画
fig,ax=plt.subplots()
ax.plot(t,s)

#グラフaxにラベルを設定する
ax.set(xlabel='time(s)',ylabel='voltage(mV)',title='About as simple as gets,folks')
#グラフaxにグリッドをつける
ax.grid()
#pyplotで表示
#ax.plot(t,s)
plt.show()
```
